```python
import pandas as pd
import requests
import json
from pandas import json_normalize

# 1. CARGAR DATOS DESDE URL
def cargar_json(url):
    """Carga JSON desde URL"""
    response = requests.get(url)
    data = response.json()
    print(f"âœ… Datos cargados: {len(data) if isinstance(data, list) else 1} registros")
    return data

# 2. NORMALIZAR JSON
def normalizar_json(data):
    """Normaliza JSON anidado a DataFrame"""
    if isinstance(data, list):
        df = json_normalize(data)
    else:
        df = json_normalize([data])

    # Limpiar nombres de columnas
    df.columns = df.columns.str.replace('.', '_', regex=False)
    df.columns = df.columns.str.lower()

    print(f"âœ… DataFrame creado: {df.shape[0]} filas Ã— {df.shape[1]} columnas")
    return df

# 3. EJECUTAR TODO DE UNA VEZ
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Cargar y normalizar
data = cargar_json(url)
df = normalizar_json(data)

# Ver resultado
print(f"\nðŸ“Š RESULTADO:")
print(f"Columnas: {list(df.columns)}")
print(f"Columnas: {list(df.columns)}")
print(f"\nPrimeras 3 filas:")
display(df.head(3))

# Guardar CSV (opcional)
df.to_csv('datos_normalizados.csv', index=False)
print(f"\nðŸ’¾ Archivo guardado: datos_normalizados.csv")

# Para descargar en Colab
from google.colab import files
files.download('datos_normalizados.csv')
```
